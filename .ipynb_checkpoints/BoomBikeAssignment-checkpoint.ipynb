#### Problem Statement:
A US bike-sharing provider BoomBikes have contracted a consulting company to understand the factors on which the demand for these shared bikes depends. Specifically, they want to understand the factors affecting the demand for these shared bikes in the American market. The company wants to know:
Which variables are significant in predicting the demand for shared bikes.
How well those variables describe the bike demands

Business Goal:
You are required to model the demand for shared bikes with the available independent variables. It will be used by the management to understand how exactly the demands vary with different features. They can accordingly manipulate the business strategy to meet the demand levels and meet the customer's expectations. Further, the model will be a good way for management to understand the demand dynamics of a new market.

### Splitting the Data into Training and Testing Sets.
The basic step for any regression model, is to create Training and Testing data sets

### Rescaling the features
Based on the above data the numerical features have different set of values, if we try to use this as its, the co-efficent value may vary and will skew the model, hence all the numeric variables needed to be rescaled. The Method which is widely used is min-max scalling.
    <br> min-max scaling = (x-xmin)/(xmax-xmin)

```python
bb_train.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
# Apply scaler() to all numeric columns 
num_vars=['atemp','hum','windspeed','cnt']
bb_train[num_vars]=scaler.fit_transform(bb_train[num_vars])
```

### Building a linear model
Recursive Feature Elimination

```python
# importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of variable equal to 15
lm=LinearRegression()
lm.fit(X_train,y_train)
rfe=RFE(lm,15)
rfe=rfe.fit(X_train,y_train)
```

```python
col_rfesupport=X_train.columns[rfe.support_]
col_rfesupport
```

```python
X_train.columns[~rfe.support_]
```

#### Building Model using statsmodel, for detail statistics

```python
import statsmodels.api as sm
def createAndCheckModel(X_trn,y_trn):
    X_trn=sm.add_constant(X_trn)
    #Linear Model and fitting it
    lm=sm.OLS(y_trn,X_trn).fit()
    return lm
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe=X_train[col_rfesupport]
```

```python
# Adding a constant variable

#X_train_rfe=sm.add_constant(X_train_rfe)
```

```python
#Running Linear model
lm1=createAndCheckModel(X_train_rfe,y_train)
print(lm1.summary())
```

```python
X_train_rfe.columns
```

#### Checking VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Create a VIF Function
def checkVIF(X_trn):
    vif=pd.DataFrame()
    X_vif=X_trn
    vif['Features']=X_vif.columns
    vif['VIF'] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif
```

```python
X_train_rfe.columns
```

```python
print(checkVIF(X_train_rfe))
```

```python
# droping one variable at time workingday as it have high p value
X_train_rfe_v1=X_train_rfe.drop(['holiday'],axis=1)
```

```python
lm2=createAndCheckModel(X_train_rfe_v1,y_train)
print(lm2.summary())
```

```python
print(checkVIF(X_train_rfe_v1))
```

```python
X_train_rfe_v2=X_train_rfe.drop(['holiday','sunday'],axis=1) # removing sunday as it have high p value
lm2=createAndCheckModel(X_train_rfe_v2,y_train)
print(lm2.summary())
print(checkVIF(X_train_rfe_v2))
```

```python
X_train_rfe_v3=X_train_rfe.drop(['holiday','sunday','hum'],axis=1) # removing one more column hum, as it is negative corelated coefficent 
lm3=createAndCheckModel(X_train_rfe_v3,y_train)
print(lm3.summary())
print(checkVIF(X_train_rfe_v3))
```

```python
X_train_rfe_v3.columns
```

```python
X_train_rfe_v4=X_train_rfe.drop(['holiday','sunday','hum','workingday'],axis=1) # removing workingday which is not highly coorelated to dependent variable where as atemp is highly coorelated to cnt
lm4=createAndCheckModel(X_train_rfe_v4,y_train)
print(lm4.summary())
print(checkVIF(X_train_rfe_v4))
```

```python
X_train_rfe_v5=X_train_rfe.drop(['holiday','sunday','hum','workingday','windspeed'],axis=1) # removing one more variable windspeed 
lm5=createAndCheckModel(X_train_rfe_v5,y_train)
print(lm5.summary())
print(checkVIF(X_train_rfe_v5))
```

```python
X_train_rfe_v6=X_train_rfe.drop(['hum','july','workingday','sat','summer','windspeed'],axis=1) # removing one july as its high VIF
lm6=createAndCheckModel(X_train_rfe_v6,y_train)
print(lm6.summary())
print(checkVIF(X_train_rfe_v6))
```

Based on iterative model building process 
lm6 model seems to be one of the best fit model having low value of VIF and p-value for all the features, with R2 and Adjusted R2 of 82.2% & 81.9% respecively.
